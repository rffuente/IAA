```python
import numpy as np

%matplotlib inline

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import widgets
from ipywidgets import IntSlider
```

```python
def boxplot_iterations(iterations):
    N = len(iterations)
    fig, ax = plt.subplots(1, N, figsize=(20, 8), sharey=True)
    for k in range(N):
        t = iterations[k]
        ax[k].boxplot(t)
        ax[k].set_title("K = " + str((k+1)*10), fontsize = 20)
        ax[k].grid(True)
        ax[k].tick_params(axis='x', which='major', labelsize=20)
        ax[k].tick_params(axis='y', which='major', labelsize=20)
    plt.tight_layout()
    plt.suptitle("Boxplots de iteraciones para distintos centroides K", fontsize = 20)
    plt.subplots_adjust(top=0.85)
    plt.show()
```

```python
def boxplot(times):
    N = len(times)
    fig, ax = plt.subplots(1, N, figsize=(20, 8))
    for k in range(N):
        t = list(times[k])
        ax[k].boxplot(t)
        ax[k].set_title("NL" + str((k+4)*2), fontsize = 20)
        ax[k].grid(True)
        ax[k].tick_params(axis='x', which='major', labelsize=20)
        ax[k].tick_params(axis='y', which='major', labelsize=20)  
        ax[k].set_xlabel(r"$K$", fontsize = 20)
    plt.tight_layout()
    ax[0].set_ylabel(r"Funci√≥n objetivo", fontsize = 20)
    #plt.suptitle(r"Boxplots de Fitness Fu", fontsize = 20)
    plt.subplots_adjust(top=0.85)
    plt.savefig("boxplot.pdf",bbox_inches = "tight", dpi=300)
    #plt.show()
```
